# Math
Eksempler pÃ¥ funksjoner vi kan bruke: 
- math.pi() 
- math.ceil(x) - *runder x oppover* 
- math.floor(x) - *runder x nedover* 
- math.sqrt(x) - *tar kvadratroten av x* 
- math.log(x) - *logaritmen av x* 
- math.cos(x), math.sin(x), math.tan(X) - *cos, sin og tan av x* 
- math.comb(n, k) - *velge k items fra n items uten repetisjon og orden*
Fullstendig liste: https://docs.python.org/3/library/math.html

```python
import math

x = 3.9 
y = 16

print(f"Hente ut ð: {math.pi:.2f}")
print(f"Tall: {x}, avrundet oppover: {math.ceil(x)}")
print(f"Tall: {x}, avrundet nedover: {math.floor(x)}")
print(f"Kvadradroten av {y} = {math.sqrt(y)}")
print(f"Logaritmen av {y} = {math.log(y):.2f}")
print(f"cos({y}) = {math.cos(y):.2f}")
print(f"{math.comb(2, 1)}")

```

# Random

Innebygd modul for Ã¥ lage tilfeldige tall. Ofte bruker vi randint() funksjonen. 

For mer omfattende dokumentasjon, se: https://docs.python.org/3/library/random.html

```python
import random 

# Simulere et terningkast 
def roll_dice():
    return random.randint(1,6)
    
print(f"Kastet en terning og fikk {roll_dice()} \n\n")



# Lage en enkel AI mot AI -> Stein, papir, saks: 
choices = ["stein", "papir", "saks"]
ai1_choice, ai2_choice = choices[random.randint(0,2)], choices[random.randint(0,2)]
print(f"Spiller 1: {ai1_choice}, Spiller 2: {ai2_choice}")
```

# NumPy
- Bibliotek for Ã¥ jobbe med arrays
- Samme formÃ¥l som vanlige lister/arrays i Python, men mye raskere 



```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6]) # eller: arr = np.arange(1, 7) 

print(f"Array = {arr}")
print()

print(f"Reshape {arr} fra 1D til 2D: {arr.reshape(2,3)}")
```

```python
print("Iterere over et NumPy-array")

for x in arr:
    print(x)
    
print()
print("Vi kan lage filter for arrayet:")
print("Setter alle verdier mindre enn 3 til Ã¥ vÃ¦re 0")

arr[arr < 3] = 0

print(arr)
```

# Matplotlib 
- Low-level grafplottingsverktÃ¸y for Python
- **OBS**: David lager alltid eksamensoppgave med plotting 

Tips: https://matplotlib.org/stable/gallery/index.html her finner dere mange eksempel pÃ¥ ulike plot


```python
import matplotlib.pyplot as plt 
from math import sin, cos

xs = [n / 10 for n in range(101)]
ys_1 = [cos(x) for x in xs]
ys_2 = [-cos(x) for x in xs]

plt.plot(xs, ys_1, "-.b") # plotter grafen til xs og ys_1 med mÃ¸nsteret: -. i fargen blÃ¥ 
plt.plot(xs, ys_2, "-r") # plotter grafen til xs og ys_2 med mÃ¸nsteret: - (vanlig linje) i fargen rÃ¸d 

plt.title("ÃndrÃ¨s hjemmesnekra Matplotlib kunst") # setter hovedtittel pÃ¥ plottet

plt.xlabel("x") # navngir x-akse
plt.ylabel("y") # navngir y-akse

plt.legend(["kul graf", "kul graf motsatt vei"]) # Lager boks som viser hva grafene viser til. PS: husk rekkefÃ¸lge hvis du har flere grafer
plt.savefig("kulgraf.png")
#plt.show() # Viser grafen i interaktivt vindu
```
